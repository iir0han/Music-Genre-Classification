```python
#import modules and libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob #for reading files
import os
import librosa as lr
```

```python
#set directory for source files
data_dir= 'C:\\Users\\rajki\\Downloads\\Datasets\\Heartbeat datasets\\set_a\\'
audio_files=glob(data_dir + '/*.wav') #glob modules reads retrives files that matches a specific pattern

len(audio_files)
```

```python
#Read the first audio file and create time array(timeline)
audio, sfreq = lr.load(audio_files[69]) #here sfreq means frequency of audio files in seconds
time=np.arange(0,len(audio))/sfreq
```

```python
time
```

```python
#plot audio over time
fig, ax=plt.subplots()
ax.plot(time,audio)
ax.set(xlabel = 'Time in s', ylabel='Sound Amplitude' )
plt.show()
```

```python
%%time #this command calculates the time required to run this block (cell) of codes

#Read all the audio files & plot their time-amplitude axes
for file in range(0, len(audio_files), 1):
    audio, sfreq = lr.load(audio_files[file]) #here sfreq means frequency of audio files in seconds
    time=np.arange(0,len(audio))/sfreq
    
    fig, ax=plt.subplots()
    ax.plot(time,audio)
    ax.set(xlabel = 'Time in s', ylabel='Sound Amplitude' )
    plt.show()
    
    
```
